<h1>Converting Categorical Features</h1>
<p>Categorical features must convert into numerical features, because model do not response for categorical features. Label encoding is used.</p>

```python
#getting unique values in Species column
train_data['Species'].unique()
```

```python
species = ('Iris-setosa', 'Iris-versicolor', 'Iris-virginica')
```

```python
#converting  type of columns to category
train_data['Species'] = train_data['Species'].astype('category')
```

```python
train_data.head()
```

```python
#assigning numerical values storing into new column
train_data['Species_Cat'] = train_data['Species'].cat.codes
```

```python
train_data['Species_Cat'].unique()
```

```python
train_data.head()
```

```python
#removing Species column
train_data.drop('Species',axis=1,inplace=True)
```

```python
train_data.head()
```

<hr style="border:10px solid #A34BB7 "> </hr>
<h1>Splitting Dataframe</h1>

```python
#importing train_test_split statement from sklearn.model_delection module
from sklearn.model_selection import train_test_split
```

```python
#split dataset as training and testing data.
x_train, x_test, y_train, y_test = train_test_split(train_data.drop('Species_Cat',axis=1), 
                                                    train_data['Species_Cat'], test_size=0.30, 
                                                    random_state=101)
```

<hr style="border:10px solid #A34BB7 "> </hr>
<h1>Training the Model</h1>

```python
#importing KNeighborsClassifier model
from sklearn.neighbors import KNeighborsClassifier
```

```python
#creating the model and fit data into the model
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
```

```python
#predicting values for test data set
predictions = neigh.predict(x_test)
```

<hr style="border:10px solid #A34BB7 "> </hr>
<h1>Model Evaluation</h1>

```python
#importing confusing matrix
from sklearn.metrics import confusion_matrix
```

```python
#evaluating confusion matrix for the model
accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
#importing accuracy_score
from sklearn.metrics import accuracy_score
```

```python
#evaluating accuracy_score
accuracy=accuracy_score(y_test,predictions)
```

```python
accuracy
```

```python
#importing evaluation report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
